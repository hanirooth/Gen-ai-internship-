```python
#Problem 1: Mobile Recharge Validation System


def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount >= 50 and amount in valid_plans:
        return True
    return False


while True:
    amt = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amt):
        print("Recharge Successful ")
        break
    else:
        print("Invalid Recharge. Try again ")



#Problem 2: Inventory Reorder Alert System


def inventory_alert(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")


products = {
    "Soap": 10,
    "Rice": 25,
    "Oil": 12,
    "Sugar": 20
}

inventory_alert(products)



#Problem 3: Student Result Processing System


def student_result(marks):
    total = 0

    for m in marks:
        total += m

    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"


marks = [60, 45, 70, 40]
print("Result:", student_result(marks))



#Problem 4: Cab Fare Estimator with Retry Option


def calculate_fare(km, peak):
    fare = 50 + (12 * km)

    if peak == "yes":
        fare += fare * 0.25

    return fare


while True:
    km = float(input("Enter distance in KM: "))
    peak = input("Peak hour? (yes/no): ").lower()

    total = calculate_fare(km, peak)
    print("Total Fare: ₹", total)

    retry = input("Retry? (yes/no): ")
    if retry != "yes":
        break



#Problem 5: Employee Attendance Eligibility Checker


def attendance_check(attendance):
    present = 0

    for a in attendance:
        if a == "P":
            present += 1

    percent = (present / len(attendance)) * 100

    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P","A","P","P","P","A","P"]
print(attendance_check(attendance))




#Problem 6: Password Strength Checker


def check_password(password):
    has_digit = False
    has_special = False
    special = "@#$"

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


pwd = input("Enter password: ")
print(check_password(pwd))









```
