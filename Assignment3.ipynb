```python
#Problem Statement 1: Employee Performance Bonus Eligibility


employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
max_score = max(employees.values())

# Find all employees with highest score
top_performers = [name for name, score in employees.items() if score == max_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")


# Problem 2: Search Query Keyword Analysis
# ================================

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase and remove punctuation
query = query.lower()
query = query.translate(str.maketrans("", "", string.punctuation))

words = query.split()

# Count frequency
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

# Display only words appearing more than once
result = {k: v for k, v in freq.items() if v > 1}

print("\nRepeated Keywords:")
print(result)


# ================================
# Problem 3: Sensor Data Validation
# ================================

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

print("\nValid Sensor Readings (Hour, Value):")
print(valid_readings)


# ================================
# Problem 4: Email Domain Usage Analysis
# ================================

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

print("\nEmail Domain Usage:")
for domain, count in domain_count.items():
    percentage = round((count / total) * 100)
    print(f"{domain}: {percentage}%")


# ================================
# Problem 5: Sales Spike Detection
# ================================

sales = [1200, 1500, 900, 2200, 1400, 3000]

average = sum(sales) / len(sales)
threshold = average * 1.3   # 30% above average

print("\nSales Spikes:")
for i, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {i}: {value}")


# ================================
# Problem 6: Duplicate User ID Detection
# ================================

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

print("\nDuplicate User IDs:")
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")



```
